```python
import numpy as np
import matplotlib.pyplot as plt

# Ma'lumotlar
x = np.array([1, 2, 3, 4, 5])  # Reklama byudjeti (ming so'm)
y = np.array([5, 7, 9, 11, 13])  # Sotuvlar (dona)

# Chiziqli regressiya (y = mx + b) ni hisoblash
# numpy.polyfit funksiyasi linearlik regressiyani hisoblaydi
m, b = np.polyfit(x, y, 1)

# Modelga asoslangan prognozlar
y_pred = m * x + b

# Natijalarni ekranga chiqarish
print(f"Regrssiya tenglamasi: y = {m:.2f}x + {b:.2f}")

# Grafikni chizish
plt.scatter(x, y, color='red', label='Ma\'lumotlar')  # Original ma'lumotlar
plt.plot(x, y_pred, label=f'Regressiya chizig\'i: y = {m:.2f}x + {b:.2f}', color='blue')  # Chiziqli regressiya
plt.xlabel('Reklama byudjeti (ming so\'m)')
plt.ylabel('Sotuvlar (dona)')
plt.title('Linear Regression')
plt.legend()
plt.show()

# X uchun prognozlar
new_x = 6
predicted_y = m * new_x + b
print(f"Reklama byudjeti {new_x} ming so\'m bo\'lsa, sotuvlar taxminan {predicted_y:.2f} dona bo\'ladi.")
```
