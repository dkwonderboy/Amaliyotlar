```python
import numpy as np
import matplotlib.pyplot as plt

# Ma'lumotlar: Harorat (x) va energiya iste'moli (y)
x = np.array([5, 10, 15, 20, 25, 30, 35])  # Harorat (C)
y = np.array([250, 230, 200, 180, 150, 120, 100])  # Energiya iste'moli (kWh)

# Polinomial regressiya (2-darajali) ni hisoblash
degree = 2  # Polinomial darajasi
coefficients = np.polyfit(x, y, degree)

# Polinomial tenglama (y = a_n * x^n + a_(n-1) * x^(n-1) + ... + a_1 * x + a_0)
polynomial = np.poly1d(coefficients)

# Modelga asoslangan prognozlar
y_pred = polynomial(x)

# Natijalarni ekranga chiqarish
print(f"Polinomial regressiya tenglamasi: y = {coefficients[0]:.2f}x^2 + {coefficients[1]:.2f}x + {coefficients[2]:.2f}")

# Grafikni chizish
plt.scatter(x, y, color='red', label='Ma\'lumotlar')  # Original ma'lumotlar
plt.plot(x, y_pred, label=f'Polinomial regressiya: y = {coefficients[0]:.2f}x^2 + {coefficients[1]:.2f}x + {coefficients[2]:.2f}', color='blue')  # Polinomial regressiya
plt.xlabel('Harorat (C)')
plt.ylabel('Energiya iste\'moli (kWh)')
plt.title('Polinomial Regression: Harorat va Energiya Iste\'moli')
plt.legend()
plt.show()

# Yangi x uchun prognozlar
new_x = 28  # Haroratni 28C deb olamiz
predicted_y = polynomial(new_x)
print(f"Harorat {new_x}C bo\'lsa, energiya iste\'moli taxminan {predicted_y:.2f} kWh bo\'ladi.")
```
