```python
import pandas as pd
import numpy as np
from sklearn.svm import SVR
import matplotlib.pyplot as plt

# Ma'lumotlar: Maydon (m²), Yoshi (yil) va Narx (ming so'm)
data = {
    'Maydon': [50, 60, 70, 80, 90, 100, 120],
    'Yosh': [5, 10, 3, 7, 2, 1, 10],
    'Narx': [400, 450, 500, 550, 600, 650, 700]
}

# DataFrame yaratish
df = pd.DataFrame(data)

# X (input) va y (target) ni ajratish
X = df[['Maydon', 'Yosh']].values  # Input (X) - Ma'lumotlar
y = df['Narx'].values  # Target (y) - Narx

# SVR (Support Vector Regression) modelini yaratish
# SVR ni yaratishda kernelni 'rbf' (radial basis function) deb belgilaymiz
model = SVR(kernel='rbf', C=100, epsilon=0.1)
model.fit(X, y)

# Modelni test qilish (yangi uy uchun prognoz qilish)
new_house = np.array([[85, 5]])  # Yangi uy (maydon = 85 m², yosh = 5 yil)
predicted_price = model.predict(new_house)

print(f"Yangi uy uchun narx: {predicted_price[0]:.2f} ming so'm")

# Modelning prognozlarini vizualizatsiya qilish (2D grafik)
plt.figure(figsize=(10, 6))
plt.scatter(df['Maydon'], df['Narx'], color='red', label='Ma\'lumotlar')  # Original ma'lumotlar
plt.scatter(new_house[0][0], predicted_price[0], color='blue', label=f'Bashorat: {predicted_price[0]:.2f} ming so\'m')  # Yangi uy narxi
plt.xlabel('Maydon (m²)')
plt.ylabel('Narx (ming so\'m)')
plt.title('SVM Regression: Uy Narxini Bashorat Qilish')
plt.legend()
plt.show()
```
