```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn import tree

# Hayvonlar haqidagi ma'lumotlar
data = {
    'Yashash_Joyi': ['Jungle', 'Savanna', 'Desert', 'Jungle', 'Savanna', 'Jungle', 'Desert', 'Savanna'],
    'Yosh': ['Katta', 'Kichik', 'Kichik', 'Katta', 'Katta', 'Kichik', 'Katta', 'Kichik'],
    'Yirikligi': ['Kichik', 'Kichik', 'Kichik', 'Katta', 'Kichik', 'Katta', 'Katta', 'Kichik'],
    'Tur': [1, 0, 0, 1, 0, 1, 0, 0]  # 1 - Yirtqich, 0 - Yirtqich emas
}

# DataFrame yaratish
df = pd.DataFrame(data)

# Kategorik o'zgaruvchilarni raqamli shaklga o'tkazish
df['Yashash_Joyi'] = df['Yashash_Joyi'].map({'Jungle': 0, 'Savanna': 1, 'Desert': 2})
df['Yosh'] = df['Yosh'].map({'Katta': 1, 'Kichik': 0})
df['Yirikligi'] = df['Yirikligi'].map({'Kichik': 0, 'Katta': 1})

# X (input) va y (target) ni ajratish
X = df[['Yashash_Joyi', 'Yosh', 'Yirikligi']]  # Input (X) - Yashash Joyi, Yosh, Yirikligi
y = df['Tur']  # Target (y) - Tur (1 yoki 0)

# Ma'lumotlarni o'rgatish va test uchun ajratish (train-test split)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Decision Tree modelini yaratish
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)

# Modelni test qilish (prognoz)
y_pred = model.predict(X_test)

# Modelning aniqligini hisoblash
accuracy = accuracy_score(y_test, y_pred)
print(f"Model aniqligi: {accuracy * 100:.2f}%")

# Qaror daraxtini vizualizatsiya qilish
plt.figure(figsize=(10, 6))
tree.plot_tree(model, feature_names=['Yashash Joyi', 'Yosh', 'Yirikligi'], class_names=['Yirtqich Emas', 'Yirtqich'], filled=True)
plt.title("Decision Tree: Hayvonlarning Turini Bashorat Qilish")
plt.show()
```
